```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r'C:\Users\Aman Sahu\OneDrive\Desktop\dl-journey\Datasets\placement.csv')
```

```python
print(df.shape)
```

```python
df.head()
```

```python
sns.scatterplot(x = df['cgpa'],y = df['resume_score'], hue = df['placed'])
#hue parameter ka kaam hota hai points ko alag-alag categories (ya groups) ke basis par different colors dena
```

```python
x = df.iloc[:,0:2]
# iloc ka matlab hota hai: integer-location based indexing 
# : before comma → sabhi rows lo.
# 0:2 after comma → column index 0 aur 1 lo
print(x)
```

```python
y = df.iloc[:,-1]
print(y)
```

```python
from sklearn.linear_model import Perceptron
# scikit-learn library se Perceptron class ko import kiya gaya hai.
p = Perceptron()
# Yahan humne Perceptron model ka ek object p banaya hai.
```

```python
p.fit(x,y) # used for train model 
# p.predict(x_test) # use for prediction
```

```python
p.coef_
#Yeh trained Perceptron model ke learned weights (coefficients) ko return karta hai.
```

```python
p.intercept_ # Trained bias (threshold)
```

```python
from mlxtend.plotting import plot_decision_regions
# plot_decision_regions ek visualization tool hai jo machine learning models ke decision boundaries ko plot karta hai
# mlxtend ka full naam hai “Machine Learning Extensions”
# Ye function 2D feature space me model ke decision regions ko plot karta hai
```

```python
plot_decision_regions(x.values, y.values, clf=p, legend=2)
# clf	Trained classifier (jaise Perceptron, SVM, Logistic Regression etc.)
```
