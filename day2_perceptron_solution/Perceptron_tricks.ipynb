```python

from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
# make_classification: Creates fake data for classification.
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(10,6))
# reates a new figure (canvas) for plotting.
# figsize=(10,6): Width = 10 inches, Height = 6 inches (bigger and clearer plot).
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
# Custom Perceptron algorithm to find weights for binary classification.
    
    X = np.insert(X,0,1,axis=1)
#Adds a bias column of 1s at index 0 in X.
# Now X becomes (100×3): [1, x1, x2] for each point.
    weights = np.ones(X.shape[1])
# Starts all weights (including bias) as 1: [1, 1, 1]    
    lr = 0.1 # Small step size for adjusting weights.
    
    for i in range(1000):
        j = np.random.randint(0,100) # Randomly pick one sample (index j from 0 to 99)
        y_hat = step(np.dot(X[j],weights)) # Calculates dot product: X[j] ⋅ weights
# Passes it to step() function (activation) to get prediction (0 or 1)        
        weights = weights + lr*(y[j]-y_hat)*X[j]
# y[j] - y_hat	Error (difference between actual and predicted)   
# lr * error * X[j]	Amount of correction to apply
# weights = weights + ...	Updates weights to reduce error    
        
    return weights[0],weights[1:]
#Returns:Bias (weights[0]),Feature weights (weights[1:])    
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
